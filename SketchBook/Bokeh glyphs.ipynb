```python
from bokeh.io import output_file, show
from bokeh.plotting import figure
from bokeh.io import output_notebook

output_notebook()

x=[1,2,3,4,5]
y=[8,6,5,2,3]

plot=figure()
plot.line(x,y, line_width=2)
plot.circle(x,y,fill_color='white', size=10)

show(plot)
```

```python
from bokeh.io import output_file, show
from bokeh.plotting import figure
from bokeh.io import output_notebook

output_notebook()

xs=[[1,1,2,2],[2,2,4],[2,2,3,3]]
ys=[[2,5,5,2],[3,5,5],[2,3,4,2]]

plot=figure()
plot.patches(xs,ys,fill_color=['red','blue','green'], line_color='white')

show(plot)
```

```python
from bokeh.io import output_file, show
from bokeh.plotting import figure
from bokeh.io import output_notebook
import numpy as np

output_notebook()

x=np.linspace(0,10,1000)
y=np.sin(x)+np.random.random(1000)*0.2
plot=figure()
plot.line(x,y)

show(plot)
```

```python
from bokeh.io import output_file, show
from bokeh.plotting import figure
from bokeh.sampledata.iris import flowers
from bokeh.io import output_notebook

output_notebook()

plot=figure()
plot.circle(flowers['petal_length'],flowers['sepal_length'],size=10)

show(plot)
```

```python
from bokeh.io import output_file, show
from bokeh.plotting import figure
from bokeh.sampledata.iris import flowers
from bokeh.models import CategoricalColorMapper
from bokeh.io import output_notebook
from bokeh.plotting import ColumnDataSource

output_notebook()

source = ColumnDataSource(flowers)

mapper = CategoricalColorMapper(
    factors=['setosa','virginica','versicolor'], 
    palette=['red','blue','green'])

plot=figure(x_axis_label='petal_lenght', y_axis_label='sepal_length')
plot.circle('petal_length','sepal_length', size=10, source=source,
           color={'field': 'species',
                 'transform':mapper})

show(plot)
```
